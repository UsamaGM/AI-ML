### Import required libraries

```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
%matplotlib inline
```

### Lets now define a second order differential equation

```python
def f(t, y):
    y1, y2 = y
    return [y2, np.sin(t) - 3*y1]
```

### Here is the initial condition

```python
y0 = [-np.pi/2, np.pi/2]
```

### Time span

```python
t_span = [0, 5]
```

### Lets now solve the system numerically

```python
sol = solve_ivp(f, t_span, y0)
```

### Lets fetch the solutions

```python
t = sol.t
y1, y2 = sol.y
```

### Finally, lets plot the system

```python
plt.plot(t, y1, label='y(t)')
plt.xlabel('t')
plt.ylabel('y(t)')
plt.title('Solution of the Second-Order ODE')
plt.grid(True)
plt.legend()
plt.show()
```
